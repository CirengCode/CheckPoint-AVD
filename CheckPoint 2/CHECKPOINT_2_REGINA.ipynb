### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

display(results_df)
```

Keberadaan outliers ini kemungkinan berasal dari aplikasi dengan rating yang sangat rendah atau sangat tinggi dibandingkan mayoritas aplikasi lainnya, sehingga lebih mencerminkan variasi penilaian pengguna.

# **Eksplorasi Data (EDA)**

## **Comparison/Perbandingan**

Aktivitas: Membandingkan rata-rata Rating antara aplikasi dengan tipe gratis dan berbayar.

Tujuan: Mengetahui apakah terdapat perbedaan kualitas aplikasi berdasarkan model bisnisnya, serta mengidentifikasi apakah aplikasi berbayar cenderung memiliki rating lebih tinggi dibandingkan aplikasi gratis.

Visualisasi: Bar Chart (Diagram Batang)

```python
perbandingan_tipe = df.groupby('Type')['Rating'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x=perbandingan_tipe.index, y=perbandingan_tipe.values, palette='viridis', hue=perbandingan_tipe.index, legend=False)
plt.title('Perbandingan Rating Aplikasi Gratis dan Berbayar')
plt.xlabel('Tipe Aplikasi')
plt.ylabel('Rata-rata Rating')
plt.xticks(rotation=0)
plt.show()
```

Berdasarkan grafik perbandingan rata-rata rating antara tipe aplikasi gratis dan berbayar, terlihat adanya perbedaan nilai rata-rata rating di antara keduanya. Aplikasi Free (gratis) memiliki rating yang lebih tinggi dari aplikasi Paid (berbayar), hal ini menunjukkan bahwa aplikasi gratis tidak selalu berarti  memiliki kualitas yang lebih rendah. Ternyata banyak aplikasi gratis yang mampu memberikan pengalaman pengguna yang baik meskipun tidak dipungut biaya langsung. Melalui analisis ini, penting untuk memahami persepsi kualitas berdasarkan model monetisasi serta dapat menjadi pertimbangan dalam menentukan strategi bisnis sebuah aplikasi.

## **Relationship/Hubungan**

Aktivitas: Menganalisis hubungan antara jumlah Reviews dan Rating aplikasi.

Tujuan: Mengetahui apakah popularitas suatu aplikasi (berdasarkan banyaknya review) berbanding lurus dengan kualitas aplikasi (rating), serta mengidentifikasi apakah terdapat hubungan antara tingkat interaksi pengguna dan kepuasan pengguna.

Visualisasi: Heatmap.

```python
# Mengubah tipe data kolom review menjadi numerik
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(data=df[['Rating','Reviews']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')

plt.title('Korelasi antara Rating dan Reviews Pada Google Play Store')
plt.show()
```

Berdasarkan hasil visualisasi heatmap diatas, diperoleh nilai korelasi sebesar 0.07 antara kategori Rating dan Reviews. Nilai ini menunjukkan hubungan positif yang sangat lemah dan mendekati nol. Hal ini mengindikasikan bahwa jumlah review yang dimiliki suatu aplikasi tidak memiliki hubungan yang signifikan dengan rating yang diperoleh. Dengan kata lain, banyaknya pengguna yang memberikan ulasan tidak secara langsung menentukan tinggi atau rendahnya penilaian aplikasi tersebut.
