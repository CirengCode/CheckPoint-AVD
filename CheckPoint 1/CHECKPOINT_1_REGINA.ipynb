Dataset Google Play Store Apps

Dateset ini merupakan kumpulan data yang berisi informasi mengenai berbagai macam aplikasi yang tersedian pada platform Google Play Store Apps. Persaingan antar pengembang semakin bertingkat seiring dengan bertambahnya jumlah aplikasi yang tersedia. Data dari dataset aplikasi Google Play Store ini, akan sangat berguna untuk pengembang aplikasi maupun perusahaan pengembang dalam mengembangkan aplikasi dan menguasai pasar.
